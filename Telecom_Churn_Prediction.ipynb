```python
#checking for outliers
df1_num=df1.select_dtypes(include=['int64','float64'])
```

```python
df1_num.head()
```

```python
df1_z=pd.DataFrame()
df1_z=pd.concat([df1_z,df1],axis=0)
```

```python
df1_z.shape
```

```python
# to calculate z_score in each column we apply a for loop 
for col in df1_num.columns:
  i='z_score_{}'.format(col)
  df1_z[i]=(df1_num[col]-df1_num[col].mean())/df1_num[col].std()
```

```python
df1_z.shape
```

```python
df1_z.head()
```

```python
df1_z_cols=df1_z.columns
print(df1_z_cols[20],df1_z_cols[23],sep=" , ")
```

```python
#now we will identify the outliers from dataframe
#for that we create a empty data frame 
z_score_outliers=pd.DataFrame()
```

```python
for i in range(20,23):
  col=df1_z_cols[i]
  z_score_outliers=pd.concat([z_score_outliers,df1_z[(df1_z[col]>=3) 
                                                   | (df1_z[col]<=-3)]],axis=0)
```

```python
x=df1.drop(columns=['Churn'])
y=df1['Churn']
```

```python
from sklearn.preprocessing import MinMaxScaler
scal=MinMaxScaler(feature_range = (0, 1))
x_scaled=scal.fit_transform(x)
```

```python
x_scaled[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=95)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
Lr = LogisticRegression(random_state = 90)
Lr.fit(x_train, y_train)
y_train_predict = Lr.predict(x_train)
y_test_predict = Lr.predict(x_test)
```

```python
y_train_acc= accuracy_score(y_train_predict,y_train)
y_test_acc= accuracy_score(y_test_predict,y_test)
print('y_test_acc :',y_test_acc,'y_train_acc :',y_train_acc)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_svm = SVC(kernel='linear', C=100)
clf_svm.fit(x_train, y_train)
y_train_predict = clf_svm.predict(x_train)
y_test_predict = clf_svm.predict(x_test)
y_train_acc= accuracy_score(y_train_predict,y_train)
y_test_acc= accuracy_score(y_test_predict,y_test)
print('y_test_acc :',y_test_acc,'y_train_acc :',y_train_acc)
```

```python
KC = KNeighborsClassifier()
KC.fit(x_train, y_train)
y_train_predict = KC.predict(x_train)
y_test_predict = KC.predict(x_test)
y_train_acc= accuracy_score(y_train_predict,y_train)
y_test_acc= accuracy_score(y_test_predict,y_test)
print('y_test_acc :',y_test_acc,'y_train_acc :',y_train_acc)
```

```python
GNB = GaussianNB()
GNB.fit(x_train, y_train)
y_train_predict = GNB.predict(x_train)
y_test_predict = GNB.predict(x_test)
y_train_acc= accuracy_score(y_train_predict,y_train)
y_test_acc= accuracy_score(y_test_predict,y_test)
print('y_test_acc :',y_test_acc,'y_train_acc :',y_train_acc)
```

```python
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
y_train_predict = DTC.predict(x_train)
y_test_predict = DTC.predict(x_test)
y_train_acc= accuracy_score(y_train_predict,y_train)
y_test_acc= accuracy_score(y_test_predict,y_test)
print('y_test_acc :',y_test_acc,'y_train_acc :',y_train_acc)
```

```python
RF=RandomForestClassifier()
RF.fit(x_train, y_train)
y_train_predict = RF.predict(x_train)
y_test_predict = RF.predict(x_test)
y_train_acc= accuracy_score(y_train_predict,y_train)
y_test_acc= accuracy_score(y_test_predict,y_test)
print('y_test_acc :',y_test_acc,'y_train_acc :',y_train_acc)
```
